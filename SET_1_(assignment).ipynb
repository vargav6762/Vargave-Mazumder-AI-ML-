```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('SET_1 (1).csv')

plt.figure(figsize=(10, 5))
plt.plot(df['month'], df['news_sold'], marker='o', color='blue', linestyle='-')
plt.title('Number of News Sold Per Month')
plt.xlabel('Month')
plt.ylabel('News Sold')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

df['sales_category'] = df['news_sold'].apply(lambda x: 'High' if x > 3000 else 'Low')
print(df)

sorted_df = df.sort_values(by='news_sold', ascending=False)
print(sorted_df)

average_sales = df['news_sold'].mean()
print(f"Average number of news sold across all months: {average_sales:.2f}")

max_sales = df['news_sold'].max()
max_months = df[df['news_sold'] == max_sales]
print(max_months)
```
